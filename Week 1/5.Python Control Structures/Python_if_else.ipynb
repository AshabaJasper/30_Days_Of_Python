## If-Else Statements in Python ğŸš¦

In Python, the `ifâ€¦elifâ€¦else` statement is used for decision making, i.e., it controls the flow of execution based on certain conditions.

1. **The if Statement**: Python's `if` statement allows you to examine the current state of a program and respond appropriately to that state.

Example:

\# Simple if statement
x = 5
if x > 0:
    print("x is positive")

2. **The else Statement**: Python's `else` statement complements the `if` statement. An `if/else` pair says: "If this expression is true, run this indented block of code. Otherwise, run this code instead."

Example:

\# if-else statement
x = -5
if x > 0:
    print("x is positive")
else:
    print("x is not positive")

3. **The elif Statement**: The `elif` statement allows you to check multiple expressions for truth and execute a block of code as soon as one of the conditions is met.

Example:

\# if-elif-else statement
x = 0
if x > 0:
    print("x is positive")
elif x == 0:
    print("x is zero")
else:
    print("x is negative")

4. **Nested if Statements**: There may be a situation when you want to check for another condition after a condition resolves to true. In such a situation, you can use the nested `if` construct.

Example:

\# Nested if statement
x = 5
if x > 0:
    if x % 2 == 0:
        print("x is a positive even number")
    else:
        print("x is a positive odd number")
else:
    print("x is not positive")
