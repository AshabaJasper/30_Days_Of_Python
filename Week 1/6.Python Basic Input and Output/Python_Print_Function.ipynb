## Python Print Functions 🖨️

Python's print() function comes with a host of features to make our output look cleaner and well-formatted.

1. **Basic Print Function**: The print() function prints the specified message to the screen.

Example:

\# Basic print function
print("Hello, Python!")

2. **Print Function with separator**: The `sep` parameter determines how multiple arguments (values) are separated.

Example:

\# Print function with separator
print("Hello", "Python", sep=",")

3. **Print Function with end parameter**: The `end` parameter determines what is printed at the end. Default is newline.

Example: 

\# Print function with end parameter
print("Hello", end=" ")
print("Python")


4. **Print Formatted String**: We can format the print output using the string's format method or f-string (formatted string literals).

Example:

\# Print formatted string
name = "Pythonista"
print(f"Hello, {name}!")

## Combining Print and Input 🎹🖨️

The print and input functions can be used together to create interactive programs.

Example:

\# Combine print and input
name = input("What's your name? ")
print(f"Hello, {name}!")
