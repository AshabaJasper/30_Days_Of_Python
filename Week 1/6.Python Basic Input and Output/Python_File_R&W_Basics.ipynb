## File Reading and Writing in Python ğŸ“–âœï¸

Python provides basic functions and methods necessary to manipulate files by default. You can do most of the file manipulation using a file object.

1. **Opening a File**: Python has a built-in function `open()` to open a file. This function returns a file object, also called a handle, as it is used to read or modify the file accordingly.

Example:

\# Open a file
file = open("testfile.txt", "r")

2. **Reading a File**: The `read()` function reads the entire file.

Example:

\# Read the file
print(file.read())

3. **Writing to a File**: The `write()` function writes any string to an open file. The file must be opened in write 'w', append 'a' or exclusive creation 'x' mode.

Example:

\# Open the file in write mode
file = open("testfile.txt", "w")
file.write("Hello World!")

4. **Closing a File**: After you're done with a file, you can use the `close()` function to free up the resources that were tied with the file.

Example:

\# Close the file
file.close()

5. **Using with Statement**: The best way to close a file is by using the `with` statement. This ensures that the file is closed when the block inside the `with` statement is exited.

Example:

\# Open and read a file
with open("testfile.txt", "r") as file:
    print(file.read())
